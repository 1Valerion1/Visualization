```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt

# увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.express as px
import plotly.graph_objs as go

init_notebook_mode(connected=True)

#для корректного отображения графиков
import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
df = pd.read_csv('Medicaid1986.csv', index_col=0)
df
```

```python
# создаем Box trace для каждого жанра из наших данных
data = []
for chi in range(0, 8):
    data.append(
        go.Box(y=df[df.children==chi].income, name='Количество детей ' + str(chi))
    )

# визуализируем данные
iplot(data, show_link = False)
```

 **Вывод**: В диапазоне от 6 - 9.5 млн находятся люди, которые не заводят детей или заводят от 1 до 3, а при доходе 8,5 - 11,5 млн $ детей от 4 до 5.

```python
# создаем Box trace для каждого жанра из наших данных
data = []
for sch in range(0, 19):
    data.append(
        go.Box(y=df[df.school==sch].income, name='Количество лет учебы ' + str(sch))
    )

# визуализируем данные
iplot(data, show_link = False)
```

```python
fig = go.Figure()
fig.add_trace(go.Pie(values=df.school.value_counts().sort_index(), labels=df.school.value_counts().sort_index().index, sort=False))
fig.update_layout(
    title='Кольцевая диаграмма для признака "количество лет учебы"',
    title_x = 0.5
)
fig.show()
```

**Вывод**: 1/5 имеет школьное образование 12 классов, выше 12 и ниже 8 классов идет существенное сокращение

```python
fig = go.Figure(data=[go.Histogram(nbinsx=30, x=df['income'])])
fig.update_layout(
    title='Распределение признака "доход"',
    title_x = 0.5
)
fig.show()
```

**Вывод**: основное количество дохода находится от 6-9млн$

```python
fig = go.Figure(data=[go.Histogram(nbinsx=15, x=df['school'])])
fig.update_layout(
    title='Распределение признака "годы учебы"',
    title_x = 0.5
)
fig.show()
```

**Вывод**: После 12 классов количество людей в поступающий в ВУЗ существенно сокращается

```python
fig = go.Figure(data=[go.Histogram(nbinsx=30, x=df['health1'])])
fig.update_layout(
    title='Распределение признака "Компонент1"',
    title_x = 0.5
)
fig.show()
fig = go.Figure(data=[go.Histogram(nbinsx=30, x=df['health2'])])
fig.update_layout(
    title='Распределение признака "Компонент2"',
    title_x = 0.5
)
fig.show()
```

**Вывод**:Компонент1 имеет больше негативной и позитивной реакции, чем компонент 2

```python
fig = go.Figure(data=[go.Histogram(nbinsx=15, x=df['age'])])
fig.update_layout(
    title='Распределение признака "возраст"',
    title_x = 0.5
)
fig.show()
```

**Вывод**: основное количеств людей от 20 -40 и 70-90 лет
