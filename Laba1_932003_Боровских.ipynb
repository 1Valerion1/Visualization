## Описание данных
## Данные об использовании Medicaid

В датасете предоставлены поперечные данные, полученные в ходе опроса потребителей Medicaid 1986 года. Данные включают две группы лиц: группу зарегистрированную в демонстрационной программе управляемого медицинского обслуживания и группу сравнения платы за услугу незарегистрированных участников из двух мест Калифорнии(округи Санта-Барбара и Вентура).

##### Датасет имеет 996 записей и 14 атрибутов. Данные являются многомерными, так как количество атрибутов больше 2-х.

#### Значения атрибутов:

1. visits: Посещения, кол-во посещений врача.
2. exposure: Воздействие,продолжительность периода наблюдения при амбулаторном лечении в днях.
3. children: Дети, общее количество детей в домохозяйстве.
4. age: Возраст, возраст испытуемого.
5. income: Доход, годовой доход домохозяйства (средний диапазон доходов в миллионах долларов США).
6. health1: 1-ый основной компонент деленный на 1000, трех состояний здоровья: функциональные ограничения, острые состояния и хронические состояния.
7. health2: 2-ой основной компонент деленный на 1000, трех состояний здоровья: функциональные ограничения, острые состояния и хронические состояния.
8. access: Доступность медицинских услуг от 0 до 1.
9. married: В браке, состоит ли этот человек в браке.
10. gender: Пол опрашиваемого.(м\ж)
11. ethnicity: Этническая принадлежность.
12. school: Количество лет, проведенных в обучении.
13. enroll: Зарегистрирован ли человек в демонстрационной программе.
14. program: Указывает на демонстрационную программу управляемого ухода: Помощь семьям с детьми-иждивенцами ("afdc") или неинституционализированный дополнительный доход по страхованию ("ssi").
